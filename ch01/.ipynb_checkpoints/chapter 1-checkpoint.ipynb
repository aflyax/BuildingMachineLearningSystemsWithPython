```python
import scipy as sp
import seaborn as sns
sns.set(context = 'poster', style = 'dark')

data = sp.genfromtxt('data/web_traffic.tsv', delimiter='\t')
print(data.shape)
```

```python
x = data[:, 0]
y = data[:, 1]
print('Number of NANs in y:', sp.sum(sp.isnan(y)))
```

```python
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]

def plot_data():
    figure()
    scatter(x, y)
    title('Web traffic over the last month')
    xlabel("Time")
    ylabel("Hits/hour")
    xticks([w*7*24 for w in range(10)],
           ['week %i'%w for w in range(10)])
    autoscale(tight = True)
#     legend()
#     grid()

plot_data()
grid()
```

```python
def error(f, x, y):
    return sp.sum((f(x)-y)**2)

f3p = sp.polyfit(x, y, 3)
print('f3p:\n', f3p)

f3 = sp.poly1d(f3p)
print('f3:\n', f3)
print('error:', error(f3, x, y))

f_x = sp.linspace(0, x[-1], 1000) #x-values for plotting the fit
plot_data()
plot(f_x, f3(f_x), linewidth=4, label = 'polyfit(3)', color='red')
# sns.set_style('dark')
legend(loc='upper left')
# grid(which='y')
```

```python
inflect_x = 3.5*7*24
x_pre = x[:inflect_x]
y_pre = y[:inflect_x]
x_post = x[inflect_x:]
y_post = y[inflect_x:]

f_pre = sp.poly1d(sp.polyfit(x_pre, y_pre, 1))
f_post = sp.poly1d(sp.polyfit(x_post, y_post, 2))

# pre_error = error(f_pre, x_pre, y_pre)
print('total error =', (error(f_pre, x_pre, y_pre) +
                        error(f_post, x_post, y_post)))

plot_data()
plot(x_pre, f_pre(x_pre), linewidth=4,
     label='polyfit(1)_pre-deflect', color='green')
plot(x_post, f_post(x_post), linewidth=4,
     label='polyfit(2)_post-deflect', color='red')
legend(loc='upper left')
```

```python
print('post function iz:\n\n', f_post)

from scipy.optimize import fsolve

# find when (in weeks) hits/hour will reach 10,000
# subtract 10,000 from f_post (making 'zero' = 10,000), solve for x,
# where f(x) = 'zero'
x_reach_100K = fsolve(f_post-100000, x0=1600)
print('\nEsspect 100,000 hits/hr on week: %f' % (x_reach_100K/(7*24)))
# divide -> weeks
```
